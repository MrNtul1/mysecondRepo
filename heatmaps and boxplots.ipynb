```python
import numpy as np
import plotly.graph_objects as go

# created a list with 12 temperatues in 5columns
temperature_data = [
    [18.5, 22.2, 23.1, 22.5, 28.3, 30.8, 32.1, 32.0, 32.5, 28.5, 24.1, 18.2],  # Year 1
    [21.3, 20.9, 22.8, 25.2, 28.0, 30.5, 31.9, 31.8, 33.3, 29.9, 23.9, 19.8],  # Year 2
    [24, 23.9, 23.7, 25.8, 23.6, 31.1, 32.2, 32.1, 31.7, 28.4, 24.7, 23.5],  # Year 3
    [17.2, 20.5, 23.3, 28.6, 28.4, 30.9, 32.0, 31.9, 34.4, 26.8, 24.3, 20.1],  # Year 4
    [20.9, 18.8, 22.5, 23.9, 27.7, 30.2, 31.7, 31.6, 33.2, 28.9, 23.8, 19.7]   # Year 5
]

# Displaying the array
print("Temperature Data:")
print(temperature_data)

# Months and years for labeling
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
years = [2020, 2021, 2022, 2023, 2024]


# Creating a box plot
# here I created a Plotly figure object, which will be used to contain the traces and layout for the plot.
fig = go.Figure()
# this loop will iterate over each month
for i in range(12):
    
    fig.add_trace(go.Box(y=temperature_data[:, i], name=months[i]))
    # here it adds a box trace to the figure
    # it relates y-values for the box plot, which are the temperatures for the current month 
    # name will correspond to loop position of month
# Here, the layout of the figure is updated to include axis labels and a title
fig.update_layout(
    title="Temperatures throughout Months for the Past Five Years",
    xaxis=dict(title="Month"),
    yaxis=dict(title="Temperature")
)
# displaying the box plot figure
fig.show()

# Creating a heatmap
# created a Plotly figure object, will insert data, 
fig = go.Figure(data=go.Heatmap( # Plotly Heatmap trace. 
                   z=temperature_data,
                   x=months,
                   y=years,
                   colorscale="Jet"))
# here I added the axes
fig.update_layout(
    title="Average Monthly Temperatures Over the Past Five Years",
    xaxis=dict(title="Month"),
    yaxis=dict(title="Year")
)
# displaying the heatmap
fig.show()

fig = go.Figure()
# This loop will iterate over each month
for i in range(12):
    fig.add_trace(go.Box(y=temperature_data[:, i], name=months[i]))
    # Here it adds a box trace to the figure
    # It relates y-values for the box plot, which are the temperatures for the current month 
    # Name will correspond to the loop position of the month

# Update layout to input the axes and title
fig.update_layout(
    title="Temperatures throughout Months for the Past Five Years",
    xaxis=dict(title="Month"),
    yaxis=dict(title="Temperature")
)

# Creating a heatmap
# Created a Plotly figure object, will insert data
fig = go.Figure(data=go.Heatmap(  # Plotly Heatmap trace
                   z=temperature_data,
                   x=months,
                   y=years,
                   colorscale="Viridis"))
# Here I added the axes
fig.update_layout(
    title="Average Monthly Temperatures Over the Past Five Years",
    xaxis=dict(title="Month"),
    yaxis=dict(title="Year")
)
# Displaying the heatmap
fig.show()
```
